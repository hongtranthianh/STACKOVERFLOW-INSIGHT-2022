# Import libraries and Load data

```python
import pandas as pd
import matplotlib.pyplot as plt
import Plot_functions as f2
%matplotlib inline
```

```python
# Load data
df_origin = pd.read_csv('./stack-overflow-developer-survey-2022/survey_results_public.csv')
```

```python
# Only analyze respondents who are either a developer by profession or
# not primarily a developer but write code sometimes as part of their work
df = df_origin[(df_origin['MainBranch']=='I am a developer by profession') | 
               (df_origin['MainBranch']=='I am not primarily a developer, but I write code sometimes as part of my work')]
```

# Method of handling missing value
- All sections in this notebook will handle missing value by removing it because the insights need to be based on fact results with descriptive statistics 

# How do developers around the world level up their coding skills?

## Methods of learning code

```python
# Remove missing value and store in a new df
LearnCode_df = df[['LearnCode']][~df['LearnCode'].isnull()]

# Get value count for each learning method
val_count_df_LearnCode = f2.value_count_func(LearnCode_df, 'LearnCode', ';')

# Rename value
val_count_df_LearnCode.iloc[:,0].replace({'School (i.e., University, College, etc)':'School',
                                            'Other online resources (e.g., videos, blogs, forum)':'Online resources (e.g., video, blog, forum)',
                                            'Hackathons (virtual or in-person)':'Hackathons',
                                            'Online Courses or Certification':'Online Courses/Certification',
                                            'Books / Physical media':'Books/Physical media',
                                            'Other (please specify):':'Others'},
                                            inplace=True)
```

```python
# Horizontal bar chart
f2.percent_barh_chart(val_count_df = val_count_df_LearnCode,
                     total_respondent = LearnCode_df.shape[0],
                     title = 'Method of learning code',
                     xlabel = 'Respondent (%)')
```

## Learn Code By Online Resources

```python
# Remove missing value and store in a new df
LearnCodeOnline_df = df[['LearnCodeOnline']][~df['LearnCodeOnline'].isnull()]

# Get value count for each learning method
val_count_df_LearnCodeOnline = f2.value_count_func(LearnCodeOnline_df, 'LearnCodeOnline', ';')

# Rename value
val_count_df_LearnCodeOnline.iloc[:,0].replace({'Coding sessions (live or recorded)':'Coding sessions',
                                                'Online challenges (e.g., daily or weekly coding challenges)':'Online challenges',
                                                'Other (Please specify):':'Others'},
                                                inplace=True)
```

```python
# Horizontal bar chart
f2.percent_barh_chart(val_count_df = val_count_df_LearnCodeOnline,
                     total_respondent = LearnCodeOnline_df.shape[0],
                     title = 'Learn Code By Online Resources',
                     xlabel = 'Respondent (%)')
```
